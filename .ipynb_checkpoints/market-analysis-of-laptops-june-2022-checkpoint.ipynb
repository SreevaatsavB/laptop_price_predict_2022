```python
# Function for encoding categorical string data into integers

def encode_to_int(arr):
    d = {}
    for p in enumerate(arr):
        if p[1] not in d:
            d[p[1]] = p[0]+1
        
    return d
```

```python
req = df1["Screen Resolution"]
max_screen_res = list(req)

for i in range(len(max_screen_res)):
    res =" ".join(re.split(r'\D+',max_screen_res[i])).split()
    if len(res)==1:
        max_screen_res[i]=int(res[0])
    else:
        max_screen_res[i]=min(int(res[0]),int(res[1]))
        
for i in range(len(max_screen_res)):
    if max_screen_res[i]>=2160:
        max_screen_res[i]=2160
    elif max_screen_res[i]>=1440:
        max_screen_res[i]=1440
    elif max_screen_res[i]>=1080:
        max_screen_res[i]=1080
    elif max_screen_res[i]>=720:
        max_screen_res[i]=720
    else:
        max_screen_res[i]=480
```

```python
def ls(s): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for ele in s: 
        str1 += ele  
    
    # return string  
    return str1 
```

### Using pandasql for some data analysis

**Limitations of Pandasql**
1. As Pandasql uses SQLite, it is subjected to all the limitations of SQLite. For example, SQLite does not implement right outer join or full outer join.
2. Pandasql performs query only, it cannot perform SQL operations such as update, insert or alter tables.

The query returns a dataframe

**The globals() function returns dictionary of variables that are created in this session such as 
df_visualize and other dataframes that are loaded till now (current cell).**

The **key** of the dictionary is the **variable name** and the **value of the dictionary** contains the **actual value of the variable.**

## Cateorical features dependancies

```python
from sklearn.feature_selection import chi2
```

```python
df_price = df1["Price"]
del df1["Price"]
```

```python
df_price = pd.DataFrame(df_price)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df, train_y, test_y = train_test_split(df1, df_price, random_state=6)
```

```python
train_categorical = train_df.copy()
train_categorical = train_categorical.drop("gpu_processor", axis = 1)
train_categorical = train_categorical.drop("Weight", axis = 1)
train_categorical = train_categorical.drop("Screen Size", axis = 1)
```

```python
train_y = train_y.astype(int)
```

```python
chi_scores_tot = []

for col in train_categorical.columns:
    chi_scores = chi2(train_categorical,train_categorical[col].astype(int))
    chi_scores_tot.append([col, chi_scores[1]])
```

```python
len(chi_scores_tot)
```

```python
# Chi square scores for each induvidual categorical variable with others

fig = plt.figure(figsize=(50,30))
for i in range(1, 15):
    
    plt.subplot(2,7,i)
    plt.title(chi_scores_tot[i-1][0])
    p_values = pd.Series(chi_scores_tot[i-1][1],index = train_categorical.columns)
    p_values.sort_values(ascending = False , inplace = True)
    p_values.plot.bar(color = "mediumslateblue")
    
    
plt.show()
```

```python
resultant = pd.DataFrame(data=[(0 for i in range(len(train_categorical.columns))) for i in range(len(train_categorical.columns))], 
                         columns=list(train_categorical.columns))

resultant.set_index(pd.Index(list(train_categorical.columns)), inplace = True)

# Finding p_value for all columns and putting them in the resultant matrix
for i in list(train_categorical.columns):
    for j in list(train_categorical.columns):
        if i != j:
            chi2_val, p_val = chi2(np.array(train_categorical[i]).reshape(-1, 1).astype(int), np.array(train_categorical[j]).reshape(-1, 1).astype(int))
            resultant.loc[i,j] = p_val
```

```python
fig = plt.figure(figsize=(20,20))
sns.heatmap(resultant, annot=True,cmap='Blues')
plt.title('Chi-Square Test Heatmap')
plt.show()
```

Here the magnitude of p-values **donot** represent the degress of dependance of the categorical features, but just tell us that 2 features are independant or not.

We choose p_value < 0.05 as the threshold.

**We could see that a lot of these categorical variables are dependant on each other with the chi-squared test (heatmap)**
